```python
import math
import random
```

```python
def test_implementation(f, num_tests=100, max_int=10000):
    for i in range(100):
        x = random.randint(0, max_int)
        y = random.randint(0, max_int)
        assert f(x, y) == x * y, f"failed test {i} with ({x}, {y})"
    return "It works!"

def _split_integer(x, loc=None):
    x = str(x)
    if loc is None:
        loc = math.ceil(len(x)/2)
    return int(x[:loc]), int(x[loc:])

def _get_num_digits(x):
    return len(str(x))

def iterate_over_digits(x):
    for d in str(x)[::-1]:
        yield int(d)
```

```python
A = 3141592653589793238462643383279502884197169399375105820974944592
B = 2718281828459045235360287471352662497757247093699959574966967627
```

```python
def grade_school(x, y):
    r = 0
    for j, dy in enumerate(iterate_over_digits(y)):
        r_ = 0
        for i, dx in enumerate(iterate_over_digits(x)):
            r_ += (10 ** i) * dy * dx
        r += (10 ** j) * r_
    return r
```

```python
test_implementation(grade_school)
```

```python
def rec_int_mult(x, y, debug=False):
    nx = _get_num_digits(x)
    ny = _get_num_digits(y)
    if debug:
        print(f"Number of digits: {nx}, {ny}")
    n = ny
    if nx == 1 or ny == 1:
        return grade_school(x, y)
    else:
        a, b = _split_integer(x)
        c, d = _split_integer(y)
        if debug:
            print(f"a, b, c, d: {a},{b},{c},{d}")
            
        #ac = rec_int_mult(a, c)
        ac = grade_school(a, c)
        ad = rec_int_mult(a, d)
        #ad = grade_school(a, d)
        #bc = rec_int_mult(b, c)
        bc = grade_school(b, c)
        #bd = rec_int_mult(b, d)
        bd = grade_school(b, d)
        
        r1 = grade_school(10**math.floor((nx + ny) / 2), ac)
        r2 = grade_school(10**math.floor(nx / 2), ad)
        r3 = grade_school(10**math.floor(ny / 2),  bc)
        return r1 + r2 + r3 + bd
```

```python
def karatsuba(x, y):
    #print(x, y)
    nx = _get_num_digits(x)
    ny = _get_num_digits(y)
    n = ny
    if nx == 1 or ny == 1:
        return x * y
    else:
        a, b = _split_integer(x)
        c, d = _split_integer(y)
        p, q = a + b, c + d
        #print(f"about to compute ac: {a}*{c}")
        ac = karatsuba(a, c)
        #print(f"about to compute bd: {b}*{d}")
        bd = karatsuba(b, d)
        #print(f"about to compute pq: {p}*{q}")
        pq = karatsuba(p, q)
        adbc = pq - ac - bd
        return 10**n * ac + 10**(n //2) * adbc + bd
```
